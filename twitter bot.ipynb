```python
#modulws required for the twitter api bot
import tweepy
from time import sleep
from credentials import *
```

```python
#copy this details from the twitter api  account
consumer_key = '------------------------------'
consumer_secret = '---------------------------------------------'
access_token = '-----------------------------------------'
access_token_secret = '-------------------------------------------'
```

```python
# Access and authorize our Twitter credentials from credentials.py
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

#tweets limit of day is 2400
#1000 dm limit
#200 tweet of user
#q is used for search, q = iot means it will search for iot on twitter
for tweet in tweepy.Cursor(api.search, q=('#iot OR #FutureOfWork OR #tech  OR #robotics OR #automation  OR #AI -filter:retweets'), lang='en').items(): 
    try:
        sleep(20)
        # Add \n escape character to print() to organize tweets
        print('\nTweet by: @' + tweet.user.screen_name)

        # Retweet tweets as they are found
        tweet.retweet()
        print('Retweeted the tweet')

    except tweepy.TweepError as e:
        print(e.reason)

    except StopIteration:
        break
```
